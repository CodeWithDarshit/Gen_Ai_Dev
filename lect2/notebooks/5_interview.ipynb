```python
from dotenv import load_dotenv

load_dotenv()

from langchain_google_genai import ChatGoogleGenerativeAI


llm = ChatGoogleGenerativeAI(model = "gemini-2.5-flash-lite")

prompts = [
  ("system","tell me how can i help you"),
  ("user", "can you solve this problme , give me two problme google interview quetion and answer system design and crack the google"),

]

res = llm.invoke(prompts)

print(res.content)
```
